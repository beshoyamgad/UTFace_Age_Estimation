```

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA()
x_train = lda.fit_transform(x_train, y_train)
x_test = lda.transform(x_test)
```

```
print(f'train images: {x_train.shape}')
print(f'train labels: {y_train.shape}')
print(f' test images: {x_test.shape}')
print(f' test labels: {y_test.shape}')
```

```
sess = tf.Session()
y_train = sess.run(tf.one_hot(y_train , 116))
y_test = sess.run(tf.one_hot(y_test, 116 ))
```

```

learning_rate = 0.01
Num_iteration = 5000
```

```

X = tf.placeholder(tf.float32, [None, 14])
Y = tf.placeholder(tf.float32, [None, 116])


W = tf.Variable(.1 * np.random.randn( 14, 116).astype(np.float32))
B = tf.Variable(.1 * np.random.randn(116).astype(np.float32))
```

```
pred = tf.nn.softmax(tf.add(tf.matmul(X,W), B))
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(pred), axis=1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```
from sklearn.metrics import confusion_matrix
```

```


with tf.Session() as sesh:
    sesh.run(tf.global_variables_initializer())
    
    for i in range(Num_iteration):
   
       sesh.run(optimizer, feed_dict={X: x_train, Y:y_train})
       c = sesh.run(cost, feed_dict={X:x_train, Y:y_train})
       print(f'Num_iteration:{i:2d} cost={c:.4f}')

       correct_pred = tf.equal(tf.argmax(pred, 1), tf.argmax(Y, 1))
       accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
       acc = accuracy.eval({X: x_train, Y: y_train})
       acc_2 = accuracy.eval({X: x_test, Y: y_test})
       print(f'Train Accuracy: {acc * 100:.2f}%')
       print(f'Test Accuracy: {acc_2 * 100:.2f}%')
   
    y_pred=tf.argmax(pred, 1)
    y_true=tf.argmax(Y, 1)  

    y_predd= y_pred.eval({X: x_train, Y: y_train}) 
    y_truee= y_true.eval( {Y: y_train}) 
    z=confusion_matrix(y_truee, y_predd)
    
    #print(z)
    correct_pred = tf.equal(tf.argmax(pred, 1), tf.argmax(Y, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
    acc = accuracy.eval({X: x_train, Y: y_train})
    acc_2 = accuracy.eval({X: x_test, Y: y_test})
    print(f'Train Accuracy: {acc * 100:.2f}%'  )
    print(f'Test Accuracy: {acc_2 * 100:.2f}%' )
  
```
